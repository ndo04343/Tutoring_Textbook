## 4. Simple HTTP Server in Python

```python
import http.server
from urllib.parse import urlparse
 
class MyHandler(http.server.BaseHTTPRequestHandler):
    def do_GET(self):
        parsed_path = urlparse(self.path)
        body_parts = ['Here is body part...',
                     'Client address : {0}'.format(self.client_address),
                     'Client string : {0}'.format(self.address_string()),
                     'Command : {0}'.format(self.command),
                     'Path : {0}'.format(self.path),
                     'real path : {0}'.format(parsed_path.path),
                     'query : {0}'.format(parsed_path.query),
                     'request version : {0}'.format(self.request_version),
                     'server_version : {0}'.format(self.server_version),
                     'sys_version : {0}'.format(self.sys_version),
                     'protocol_version : {0}'.format(self.protocol_version)]
        message='\n'.join(body_parts)
        self.send_response(200)
        self.end_headers()
        self.wfile.write(message.encode('utf-8'))
        return None
 
server = http.server.HTTPServer(('localhost', 8080), MyHandler)
server.serve_forever()
```


## 5. Simple HTTP Client in Python

```python
import requests

def req_get(url, params=None):
    response = requests.get(url, params)
    print('< Response Header >')
    print(response.headers)
    print('Status-code : ', response.status_code)
    print('\n< Response Body >')
    print(response.content.decode('utf-8'))
    
req_get("http://127.0.0.1:8080/")
```
